```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
source = 'https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8'
df = pd.read_html(source, match='Коефіцієнт народжуваності в регіонах України', thousands=".", decimal=",")[0]
print(df)
```

### - Вивести перші рядки таблиці за допомогою методу head

```python
print(df.head(5))
```

### - Визначте кількість рядків та стовпців у датафреймі (атрибут shape)

```python
print(df.shape)
```

### - Замініть у таблиці значення "—" на значення NaN

```python
df = df.replace('—', 'NaN')
print(df)
```

### - Визначте типи всіх стовпців за допомогою dataframe.dtypes

```python
print(df.dtypes)
df.info()
```

### - Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"

```python
df=df.astype(dtype={'2014':'float64', '2019':'float64'})
print(df.dtypes)
```

### - Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

```python
percentage_nulls = df.isnull().mean() * 100
print(percentage_nulls)
```

### - Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
df=df.drop(df.index[-1])
print(df)
```

### - Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python
numeric_columns = df.select_dtypes(include=['int', 'float']).columns
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean().round(1))
print(df)
```

### - Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

```python
filtered_df = df[df['2019'] > df['2019'].mean()]
print(filtered_df[['Регіон','2019']])
```

### - У якому регіоні була найвища народжуваність у 2014 році?

```python
filtered_df = df[df['2014'] == df['2014'].max()]
print(filtered_df[['Регіон','2014']])
```

### - Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році

```python
df.plot(kind='bar', x='Регіон', y='2019', color='cadetblue')
plt.title('Стовпчикова діаграма народжуваності по регіонах у 2019 році')
plt.xlabel('Регіон')
plt.ylabel('Kоефіцієнт народжуваності')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x=df.columns[1:], y=df.loc[25, df.columns[1:]], marker='o', color='cadetblue')
plt.xlabel('Рік')
plt.ylabel('Kоефіцієнт народжуваності')
plt.title('Динаміка рівня народжуваності у Києві')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
palette = sns.color_palette("mako_r", 8)
df_long = pd.melt(df, id_vars=['Регіон'], value_vars=['1950', '1960', '1970', '1990', '2000', '2012', '2014', '2019'],
                  var_name='Рік', value_name='Kоефіцієнт народжуваності')
ax = sns.lineplot(data=df_long, x='Регіон', y='Kоефіцієнт народжуваності', hue='Рік', palette=palette)
ax.set_xticks(ax.get_xticks())
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.xlabel('Регіон')
plt.ylabel('Kоефіцієнт народжуваності')
plt.title('Коефіцієнт народжуваності в регіонах України')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
palette = sns.color_palette("mako_r", 27)
ax = sns.barplot(x='Регіон', y='2019', hue='Регіон', data=df, palette=palette, legend=False)
ax.set_xticks(ax.get_xticks())
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.xlabel('Регіон')
plt.ylabel('Kоефіцієн народжуваності у 2019 році')
plt.title('Kоефіцієн народжуваності у 2019 році за регіонами')
plt.show()
```
