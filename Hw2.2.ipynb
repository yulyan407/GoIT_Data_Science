### - Видаліть усі рядки у вихідній таблиці за допомогою методу dropna

### - Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python

```python
python_data = newdf[newdf['Мова програмування'] == 'Python']
```

### - Визначте розмір таблиці python_data за допомогою методу shape

```python
print(python_data.shape)
```

### - Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"

```python
python_data.groupby("Посада")
```

### - Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата на місяць"

```python
aggregated_data = python_data.groupby('Посада').agg({'Зарплата на місяць': ['min', 'max']})
print(aggregated_data)
```

### - Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"

```python
def fill_avg_salary(row):
    return row["Зарплата на місяць"].mean()

aggregated_data['avg'] = aggregated_data.apply(fill_avg_salary, axis=1)
print(aggregated_data)
```

### - Створіть описову статистику за допомогою методу describe для нового стовпчика

```python
print(aggregated_data['avg'].describe(include='all'))
```

```python
plt.figure(figsize=(12, 6))
palette = sns.color_palette("mako_r", 5)
ax = sns.boxplot(data=python_data, x='Посада', y='Зарплата на місяць', hue='Рівень англійської', palette=palette)
plt.xlabel('Посада')
plt.ylabel('Зарплата на місяць')
plt.title('Зарплата на місяць серед спеціалістів Python')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
ax = sns.histplot(python_data['Зарплата на місяць'], bins=30, kde=True, color='cadetblue')
plt.xlabel('Зарплата на місяць')
plt.ylabel('Частота')
plt.title('Розподіл зарплати серед спеціалістів Python')
plt.show()
```
